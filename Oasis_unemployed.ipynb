```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
```

```python
encoded=encoder.fit_transform(df.iloc[:,[0]])
enc_arr=encoded.toarray()
```

```python
df.head()
```

```python
df['Region'].nunique()
```

```python
ndf = pd.DataFrame(enc_arr, columns=[f'state_enc_{i}' for i in range(enc_arr.shape[1])], index=df.index)
```

```python
tdf=pd.concat([df, ndf], axis=1)
```

```python
tdf.head()
```

```python
tdf = tdf.drop(tdf.columns[[0,1, 2]], axis=1)
tdf.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
b = scaler.fit_transform(tdf)
b = pd.DataFrame(b, columns=tdf.columns)

b.head()
```

```python
b['state_enc_0']=b['state_enc_0']/5.042675027063654
```

```python
b.head()
```

```python
x=b.drop(b.columns[0],axis=1)
y=b.iloc[:,0]
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_tr,x_ts,y_tr,y_ts=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
clf.fit(x_tr, y_tr)
y_pred = clf.predict(x_ts)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
mse = mean_squared_error(y_ts, y_pred)
rmse = np.sqrt(mse)

r2 = r2_score(y_ts, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R-squared: {r2}")
```

Using Support Vector Regression

```python
from sklearn.svm import SVR
svm_model=SVR()
svm_model.fit(x_tr,y_tr)
```

```python
y_pred=svm_model.predict(x_ts)
mse = mean_squared_error(y_ts, y_pred)
rmse = np.sqrt(mse)

r2 = r2_score(y_ts, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R-squared: {r2}")
```

Using Random Forest Regressor


```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(max_depth=2,random_state=0)
rf.fit(x_tr,y_tr)
```

```python
y_pred=rf.predict(x_ts)
mse = mean_squared_error(y_ts, y_pred)
rmse = np.sqrt(mse)

r2 = r2_score(y_ts, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R-squared: {r2}")
```
